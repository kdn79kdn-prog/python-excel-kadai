```python
import openpyxl

# datetimeモジュールのdatetimeクラスをインポートする
from datetime import datetime

# ワークブックを作成
workbook = openpyxl.Workbook()

# 現在の日付（年月日）を取得して、変数todayに代入する
today = datetime.today()

# アクティブなシートを取得する
ws = workbook.active

# セルに値を書き込む
ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'
ws['F4'] = 'No.'
ws['F5'] = '日付'
ws['G4'] = '0001'
ws['G5'] = today.strftime('%Y/%m/%d')
ws['B15'] = '合計'
ws['B16'] = '消費税'
ws['B17'] = '税込合計'

# 表のヘッダー行を書き込む
header = ['商品名', '数量', '単価', '金額']

# 複数行のデータをリストdataに格納
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000]
]

start_row = 10
start_col = 2  # B列

# ヘッダーを書き込む（B10スタート）
for col, value in enumerate(header):
    ws.cell(row=start_row, column=start_col + col, value=value)

# データを書き込む（B11以降）
for row, record in enumerate(data):
    for col, value in enumerate(record):
        ws.cell(
            row=start_row + 1 + row,
            column=start_col + col,
            value=value
        )
# E13：E11とE12の合計
ws['E13'] = '=SUM(E11:E12)'

# E15：E11とE12の合計
ws['E15'] = '=SUM(E11:E12)'

# E16：E15の10％
ws['E16'] = '=E15*0.1'

# E17：E15とE16の合計
ws['E17'] = '=E15+E16'

# 日付のtodayオブジェクトを、特定のフォーマットで文字列に変換する
backup_filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

workbook.save(backup_filename)

```
